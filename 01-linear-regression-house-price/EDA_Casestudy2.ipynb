```python
#outliers
for i in df.columns:
    if df[i].dtype != 'object':
        sns.boxplot(df[i])
        plt.title(i)
        plt.show()
```

```python
df.shape
```

```python
for i in df.columns:
    if df[i].dtype != 'object':
        Q1 = df[i].quantile(0.25)
        Q3 = df[i].quantile(0.75)
        IQR = Q3 - Q1
        ll = Q1-1.5*IQR
        UL = Q3+1.5*IQR
        df = df[(df[i] <= UL) & (df[i] >= ll)]
```
